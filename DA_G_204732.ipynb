```python
#Item 1: Postitives and Negatives

number_list=[]
while True:
    number=input("Enter a number: ")
    if number=="a":
        break
    number_list.append(number)
positives=0
negatives=0
zeroes=0
for i in number_list:
    if int(i) >0:
        positives+=1
    elif int(i)<0:
        negatives+=1
    else:
        zeroes+=1

print(f"Positives - {positives}")
print(f"Negatives - {negatives}")
print(f"Zeroes - {zeroes}")
```

```python
#Item 2: The Four-Six Pour Over Coffee Brewing Method
amount_coffee=int(input("Enter amount of coffee in grams: "))


amount_water=15*amount_coffee
first_part=0.4*amount_water
second_part=0.6*amount_water

while True:
    sweet_acidic=input("Do you want sweet or acidic?(type sweet or acidic) ")
    if sweet_acidic=="sweet":
        pour_1=0.4*first_part
        pour_2=0.6*first_part
        break
    elif sweet_acidic=="acidic":
        pour_1=0.6*first_part
        pour_2=0.4*first_part
        break
    else:
        print("You have entered an invalid option. Please try again.")
        

while True:
    strength=input("Do you want your coffee strong? (Y or N)").upper()
    if strength =="Y":
        pour_second=second_part/4
        break
    elif strength=="N":
        pour_second=second_part/2
        break
    else:
        print("You have entered an invalid option. Please try again.")


print("Water pour plan: ")
print(f"First Part ({first_part})g:")
print(f"Pour #1: {pour_1}grams")
print(f"Pour #2: {pour_2}grams")
print(f"Second Part ({second_part})g:")
print(f"Pour #3: {pour_second}grams")
print(f"Pour #4: {pour_second}grams")

if strength=="Y":
    print(f"Pour #5: {pour_second}grams")
    print(f"Pour #6: {pour_second}grams")
print(f"Total water: {amount_water} g")

```

```python
#Item 3: Restaurant Cashier Simulation

def cashier():
    print("Welcome here is our menu!\n\n  - Chicken Burger: Php 60\n  - Sisig Burrito: Php 70\n  - Beef Kebab Bowl: Php 80\n  - Calamansi Juice: Php 20")
    menu={"chicken burger":60, "sisig burrito":70, "beef kebab bowl":80, "calamansi juice":20}
    total=0
    counter=0
    while counter<5:
        print("\nWhat will you have?")
        choice=input('("chicken burger", "sisig burrito", "beef kebab bowl", "calamansi juice", or "/checkout" to checkout now)\n').lower()

        if choice=="/checkout":
            break

        elif choice not in menu:
            print("Ooops, that's not on our menu!")
            continue

        else:
            total+=menu[choice]
            print(f"That's Php {total} so far.")
            counter+=1
    print(f'\nThat will be a total of {total}.')        
    payment=float(input("How much cash will you pay?\n"))

    if payment>total:
        print('\n'+f"Thank you for your payment! Your change is Php {payment-total}. Enjoy your order!")
    elif payment==total:
         print("\nThank you for giving exact amount. Enjoy your order!")
    else:
        print("\nInsufficient payment. Transaction could not be completed.")
        
cashier()
```
